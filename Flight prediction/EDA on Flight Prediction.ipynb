One approach to solve this problem can be label encoding where we will assign a numerical value to these labels for example Male and Female mapped to 0 and 1. But this can add bias in our model as it will start giving higher preference to the Female parameter as 1>0 but ideally, both labels are equally important in the dataset. To deal with this issue we will use the One Hot Encoding technique.



Label encoding can be used where we want to show hiearchy or some high to low values 
i.e.
1. phd-5
2. Be-3
3. masters-4
4. 12-2
5. 10-1m


One Hot Encoding, the categorical parameters will prepare separate columns for both Male and Female labels. So, wherever there is a Male, the value will be 1 in the Male column and 0 in the Female column, and vice-versa

# Approachs to use One hot encoding :
1. pd.get_dummies
2. sklearn Ohe library
